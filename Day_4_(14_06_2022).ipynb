```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x=linspace(-2,4,20)
plt.plot(x,sin(x))
```

```python
plt.plot(x,sin(x),'r',linewidth='1')
plt.plot(x,sin(x),'g',linewidth='3')
```

```python
plt.plot(x,-x*x+4*x-5,'r',linewidth='1')
plt.title("Parabolic Function of -x^2+4x-5")
```

```python
plt.plot(x,-x*x+4*x-5,'r',linewidth='1')
plt.title(r"Parabolic Function of $-x^2+4x-5$")
plt.xlabel(r'$x$')
plt.xlabel(r'$y$')
plt.annotate("local maxima",xy=(2,-1))
plt.xlim(-4,5)
plt.ylim(-15,2)
plt.annotate("root",xy=(-4,0))
```
